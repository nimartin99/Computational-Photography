# 1.2 Simulation of Rolling Shutter Effect (1+30+5+4=40)

In the lecture, we learned that the majority of CMOS sensors use rolling shutter, which scans across a scene rapidly rather than taking a snapshot of the instant moment. Therefore images with specific  distortions are captured when recording fast-moving objects. This is refered to as the rolling shutter distortion.  In this exercise, we would like to simulate an image with this effect by sampling a highspeed video of a rotating ceiling fan. In the uploaded dataset, you are given 360 jpg pictures with a rotating fan (e.g.Figure 2(a)-2(c)). Sample one row of each highspeed frame to the generated image to produce a rolling shutter effect such as Figure 2(d).  






_Figure 2: Video frames of rotating ceiling fan and rolling shutter effect image_

Your Tasks:
* a) Set the paths to the image directory and result location,
* b) implement a function `generate_rolling_shutter_effect` and use the frames in the given dataset to generate an image that shows the rolling shutter effect,
* c) display the image,
* d) and save the image.



```python
# Imports
import os

import imageio
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import image
```

```python
# Define paths
# Set path to image directory here: TODO
jpg_frames_path = "images/jpg_frames/"
result_path = "images/"
```

### a)
Set the pathes to the image directory and result location here:

```python
# Helper functions
def read_frames(path):
    frames = np.zeros([360, 360, 480, 3]).astype("uint8")
    for i in range(360):
        filename = str("frame_0" + "{0:03}".format(i + 1) + ".jpg")
        # print(filename)
        filename = os.path.join(path, filename)
        frame = imageio.imread(filename)
        frames[i, :, :, :] = frame
    return frames
```

```python
# Read image frames
frames = read_frames(jpg_frames_path)
print("frame shape:", frames.shape)
display = frames[359, :, :, :].astype("uint8")
plt.imshow(display)
plt.axis("off")
plt.title("frame")
```

### Generate Rolling Shutter Effect 

### b)
Implement a function `generate_rolling_shutter_effect` and use the frames in the given dataset to generate an image that shows the rolling shutter effect here:

```python
def generate_rolling_shutter_effect():
    image = np.empty((360, 480, 3), dtype=object)
    for i, frame in enumerate(frames):
        for x in range(480):
            image[i][x] = frame[i][x]
    return image
        
image = generate_rolling_shutter_effect()
```

### c)
Display the image here:

```python
def display(data1):  # create figure
    data1 = (data1 / 255).astype("float32")

    fig = plt.figure(figsize=(10, 7))
    # setting values to rows and column variables
    rows = 2
    columns = 2

    # Adds a subplot at the 1st position
    fig.add_subplot(rows, columns, 1)
    # showing image
    plt.imshow(data1)
    plt.axis("off")
    plt.title("rolling shutter effect")
    plt.savefig(result_path + "rolling_shutter_effect")
    
display(image)
```

### d)
And save the image here:

```python
def save_image(data1):
    data1 = (data1 / 255).astype("float32")
    fig = plt.figure(figsize=(10, 7))
    # setting values to rows and column variables
    rows = 2
    columns = 2

    # Adds a subplot at the 1st position
    fig.add_subplot(rows, columns, 1)
    # showing image
    plt.imshow(data1)
    plt.axis("off")
    plt.savefig(result_path + "rolling_shutter_effect")
    
save_image(image)
```
